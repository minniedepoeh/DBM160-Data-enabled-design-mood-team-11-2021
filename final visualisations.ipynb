And now for the fun part, the location graphs

```python
#a funtion creatinf the visualisations of the different tokens
def plots_tokens(namex, namey, name):
    #scatterplot with lines between the points that change colour
    fig, ax = plt.subplots(figsize = (10, 7))
    
    #set the min, max and middle to use more often
    min_x = -6
    max_x = 6
    min_y = 0
    max_y = 12
    
    #maks the NaN values (https://stackoverflow.com/questions/14399689/matplotlib-drawing-lines-between-points-ignoring-missing-data)
    series1 = token[namex].astype(np.double)
    s1mask = np.isfinite(series1)
    series2 = token[namey].astype(np.double)
    s2mask = np.isfinite(series2)
    
    #zet limits and select the data into scatterplot
    ax.set(xlim=(min_x, max_x), ylim=(min_y, max_y))
    ax.scatter(token[namex][s1mask], token[namey][s2mask])
    
    #add the scales and a title
    ax.text(min_x, max_y, "negative", color="black", fontsize=18, ha = 'left', va = 'bottom')
    ax.text(max_x, max_y, "positive", color="black", fontsize=18, ha = 'right', va = 'bottom')
    ax.text(min_x, min_y, "low intensity", color="black", fontsize=18, ha = 'right', va = 'center')
    ax.text(min_x, max_y, "high intensity", color="black", fontsize=18, ha = 'right', va = 'center')
    ax.set_title(name, fontsize=24)
    
    #set values on x- and y-axis to nothing as to not distract
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    
    #CONSTANTS
    NPOINTS = cnt.loc[name, 'count']
    COLOR='blue'
    RESFACT= 10
    MAP='winter' # choose carefully, or color transitions will not appear smoooth
    
    #credit https://stackoverflow.com/questions/8500700/how-to-plot-a-gradient-color-line-in-matplotlib
    # Choose a color map, loop through the colors, and assign them to the color 
    # cycle. You need NPOINTS-1 colors, because you'll plot that many lines 
    # between pairs. In other words, your line is not cyclic, so there's 
    # no line from end to beginning
    cm = plt.get_cmap(MAP)
    ax.set_prop_cycle(color=[cm(1.*i/(NPOINTS-1)) for i in range(NPOINTS-1)])
    for i in range(NPOINTS-1):
        ax.plot(token[namex][s1mask][i:i+2], token[namey][s2mask][i:i+2], linewidth = 5)
        
    plt.savefig(name+'_movement.png', bbox_inches='tight')
    return plt.show()
```

```python
#one plot with all the tokens

#scatterplot with lines between the points that change colour
fig, ax = plt.subplots(figsize = (10, 7))
    
#set the min, max and middle to use more often
min_x = -6
max_x = 6
min_y = 0
max_y = 12
    
#add the scales and a title
ax.text(min_x, max_y, "negative", color="black", fontsize=18, ha = 'left', va = 'bottom')
ax.text(max_x, max_y, "positive", color="black", fontsize=18, ha = 'right', va = 'bottom')
ax.text(min_x, min_y, "low intensity", color="black", fontsize=18, ha = 'right', va = 'center')
ax.text(min_x, max_y, "high intensity", color="black", fontsize=18, ha = 'right', va = 'center')
ax.set_title('All', fontsize=24)

#set values on x- and y-axis to nothing as to not distract
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])

#maks the NaN values 
def mask(name):
    series = token[name].astype(np.double)
    smask = np.isfinite(series)
    return(smask)
    
#zet limits and select the data into scatterplot
ax.set(xlim=(min_x, max_x), ylim=(min_y, max_y))
ax.scatter(token['ax'][mask('ax')], token['ay'][mask('ay')], label = 'a', linewidths = 16)
ax.scatter(token['dx'][mask('dx')], token['dy'][mask('dy')], label = 'd', linewidths = 16)
ax.scatter(token['ex'][mask('ex')], token['ey'][mask('ey')], label = 'e', linewidths = 12)
ax.scatter(token['fx'][mask('fx')], token['fy'][mask('fy')], label = 'f', linewidths = 12)
ax.scatter(token['gx'][mask('gx')], token['gy'][mask('gy')], label = 'g', linewidths = 8)
ax.scatter(token['hx'][mask('hx')], token['hy'][mask('hy')], label = 'h', linewidths = 8)
ax.scatter(token['ix'][mask('ix')], token['iy'][mask('iy')], label = 'i', linewidths = 4)
ax.scatter(token['jx'][mask('jx')], token['jy'][mask('jy')], label = 'j', linewidths = 4)

plt.legend()
plt.savefig('all_movement.png', bbox_inches='tight')

plt.show()
```
